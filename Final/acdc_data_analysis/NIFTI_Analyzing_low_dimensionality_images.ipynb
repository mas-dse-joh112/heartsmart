```python
list_x_below_216=[]
dir_x={}
list_y_below_216=[]
dir_y={}
for root,subdirs,files in os.walk(path_images):
    #print(root)
    for file in os.listdir(root):
        filePath = os.path.join(root, file) 
        if filePath.endswith('nii.npy'):
            image=np.load(filePath)
            #print(image.shape)
            try:
                if image.shape[0]<256:
                    list_x_below_216.append(filePath)
                    dir_x[filePath]=image.shape[0]
                elif image.shape[1]<256:
                    list_y_below_216.append(filePath)
                    dir_y[filePath]=image.shape[1]
            except:
                null

        
```
