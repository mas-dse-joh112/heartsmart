```python
import dicom as dm
import glob
import pandas as pd
```

```python
L=glob.glob("/Users/egouda/Downloads/sunny_brook_data/*/*/*/*.dcm")
```

```python
Kaggle=glob.glob('/Users/egouda/Downloads/train/*/*/*/*')
```

```python
Kaggle_v=glob.glob('/Users/egouda/Downloads/validate/*/*/*/*')
```

```python
def read_dicom_filtered(filepath):
    List_of_image=[]
    c = dm.read_file(filepath)
    try:
        A=c.SliceThickness
    except AttributeError as error:
        A="N/A"
        pass
    try:
        B=c.InPlanePhaseEncodingDirection
    except AttributeError as error:
        B="N/A"
        pass
    try:
        C=c.SliceLocation
    except AttributeError as error:
        C="N/A"
        pass

    try:
        D=c.PixelSpacing
    except AttributeError as error:
        D="N/A"
        pass
    try:     
        E=c.ImageOrientationPatient
    except AttributeError as error:
        E="N/A"
        pass
    try:     
        F=c.SpacingBetweenSlices
    except AttributeError as error:
        F="N/A"
        pass
    try:     
        G=c.PatientSex
    except AttributeError as error:
        G="N/A"
        pass
    try:     
        H=c.PatientWeight
    except AttributeError as error:
        H="N/A"
        pass
    try:     
        I=c.PatientHeight
    except AttributeError as error:
        I="N/A"
        pass
    try:     
        J=c.PatientAge
    except AttributeError as error:
        J="N/A"
        pass
    
    
    
    List_of_image=[A,B,C,D,E,F,G,H,I,J]    
    return List_of_image
```

```python
listA=[]
for i in Kaggle_v :
    b=read_dicom_filtered(i)
    listA.append(b)
```

```python
listA=[]
for i in L:
    b=read_dicom_filtered(i)
    listA.append(b)
```

```python
df=pd.DataFrame(listA, columns=['SliceThickness','InPlanePhaseEncodingDirection','SliceLocation','PixelSpacing','ImageOrientationPatient','SpacingBetweenSlices','PatientSex','PatientWeight','PatientHeight','PatientAge'])
```

```python
dfx=pd.DataFrame(Kaggle_v,columns=['Path'])
```

```python
Result=pd.concat([df,dfx], axis=1)
```

```python
Result.to_csv('Merged_Path_kaggle_validate.csv')
```
