```python
performances = !/usr/bin/find /masvol/heartsmart/unet_model -name *performance.json
```

```python
dfperfs = []

index = 0

for p in performances:
    print ("|"+p+"|")
    with open (p,'r') as inputs:
        
        nodes = p.split('/')
        jin = None
        try:
            jin = json.load(inputs)
        except:
            continue
        
        stats = os.stat(p)
        print (stats)
        print (stats.st_ctime)
        datestr = datetime.datetime.fromtimestamp(int(stats.st_ctime)).strftime('%Y-%m-%d %H:%M:%S')
        print (datestr)
        jin.update({'model_name':nodes[-1].replace('.json','')})
        jin.update({'model_create':datestr})
        print (jin)
        df = pd.DataFrame(jin,index=[index])
        dfperfs.append(df)
        index += 1
        #break
```

```python
dfall = pd.concat(dfperfs,axis=0)
```
